```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Hitesh Ahuja\Downloads\heart.csv")
df.head()
```

```python
cols = df.columns.tolist()
cols
```

```python
import random

# Define parameters for the population
GENOME_LENGTH = len(cols)-  1 ;
POPULATION_SIZE = GENOME_LENGTH * 5  

GENE_VALUES = [0, 1]

# Create a function to generate a random genome
def generate_genome(length):
    genome = []
    for i in range(length):
        genome.append(random.choice(GENE_VALUES))
    return genome

# Create the initial population
population = []
for i in range(POPULATION_SIZE):
    genome = generate_genome(GENOME_LENGTH)
    population.append(genome)

# Print the initial population
print(population)
```

```python
n_tribes = 5
tribe_size = POPULATION_SIZE // n_tribes
tribes = []

# Shuffle the population to randomly distribute genomes
random.shuffle(population)

# Divide the population into tribes
for i in range(n_tribes):
    start_index = i * tribe_size
    end_index = start_index + tribe_size
    tribe = population[start_index:end_index]
    tribes.append(tribe)

# Print the tribes
for i, tribe in enumerate(tribes):
    print(f"Tribe {i+1}: {tribe}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics

# Define the fitness function
def fitness_function(decision_tree):
    # Evaluate the fitness of the decision tree
    size = len(decision_tree) ;
    X = pd.DataFrame();
    for i in range(size):
        
        if decision_tree[i] == 1 :

            X[cols[i]] = df[cols[i]] ;
            
    Y = df['target'] ;
    
    X_train , X_test , Y_train , Y_test = train_test_split(X , Y , train_size = 0.7 , test_size = 0.3 , random_state = 100)
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(X_train, Y_train)
#     tree.plot_tree(clf)
    
    y_pred = clf.predict(X_test)
    return metrics.accuracy_score(Y_test, y_pred)

```

```python
# Iterate over each tribe and evaluate the fitness of each decision tree
for i, tribe in enumerate(tribes):
    print(f"Evaluating tribe {i+1}")
    for decision_tree in tribe:
        fitness = fitness_function(decision_tree)
        print(f"  Decision tree {decision_tree}: fitness = {fitness}")
```

```python
# Define the number of champions to select from each tribe
num_champions = 3

# Iterate over each tribe and select the champions based on their fitness
for i, tribe in enumerate(tribes):
    print(f"Selecting champions for tribe {i+1}")
    # Sort the decision trees in the tribe based on their fitness
    sorted_trees = sorted(tribe, key=fitness_function, reverse=True)
    # Select the top n decision trees as champions
    champions = sorted_trees[:num_champions]
    # Print the champions and their fitness
    for decision_tree in champions:
        fitness = fitness_function(decision_tree)
        print(f"  Champion {decision_tree}: fitness = {fitness}")
    # Replace the tribe with the champions
    tribes[i] = champions
```

```python
# Define the competition function
def compete_champions(champions):
    # Evaluate the fitness of each champion
    champion_fitness = [fitness_function(champion) for champion in champions]
    # Determine the index of the champion with the highest fitness
    best_champion_idx = champion_fitness.index(max(champion_fitness))
    # Return the best champion
    return champions[best_champion_idx]

# Initialize the list of all champions
all_champions = []

# Iterate over each tribe and select the champions based on their fitness
for i, tribe in enumerate(tribes):
    print(f"Selecting champions for tribe {i+1}")
    # Sort the decision trees in the tribe based on their fitness
    sorted_trees = sorted(tribe, key=fitness_function, reverse=True)
    # Select the top n decision trees as champions
    champions = sorted_trees[:num_champions]
    # Print the champions and their fitness
    for decision_tree in champions:
        fitness = fitness_function(decision_tree)
        print(f"  Champion {decision_tree}: fitness = {fitness}")
    # Add the champions to the list of all champions
    all_champions.extend(champions)

# Compete the champions against each other and select the best champion
best_champion = compete_champions(all_champions)

# Print the best champion and its fitness
fitness = fitness_function(best_champion)
print(f"The best champion is {best_champion} with fitness {fitness}")

# print(all_champions)
```

```python
def apply_genetic_operators(champions):
    # Select two parent 
#     print(champions)
    parent1, parent2 = random.sample(champions, 2)
#     print(parent1)
    # Perform crossover to create two new solutions
    crossover_point = random.randint(1, len(parent1)-1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    
    # Perform mutation on the new solutions
    mutation_rate = 0.1
    for i in range(len(child1)):
        if random.random() < mutation_rate:
            if isinstance(child1[i], bool):
                child1[i] = not child1[i]
            elif isinstance(child1[i], int):
                child1[i] += random.randint(-1, 1)
            elif isinstance(child1[i], float):
                child1[i] += random.uniform(-1, 1)
                
    for i in range(len(child2)):
        if random.random() < mutation_rate:
            if isinstance(child2[i], bool):
                child2[i] = not child2[i]
            elif isinstance(child2[i], int):
                child2[i] += random.randint(-1, 1)
            elif isinstance(child2[i], float):
                child2[i] += random.uniform(-1, 1)
    
    # Return the new solutions
    return [child1, child2]
```

```python
def evaluate_fitness(new_solutions):
    # Evaluate the fitness of each new solution
    new_fitness = [fitness_function(solution) for solution in new_solutions]
    # Return the new solutions and their fitness values as a list of tuples
    return list(zip(new_solutions, new_fitness))


```

```python
def replace_weakest(new_solutions, tribes):
    # Iterate over each tribe
    print(new_solutions)
    print(tribes)
    
    
    for tribe in tribes:
        # Sort the decision trees in the tribe based on their fitness
        sorted_trees = sorted(tribe, key=fitness_function)
        # Replace the weakest decision trees with the new solutions
        for i in range(len(tribe)):
            tribe[i] = new_solutions[i] if fitness_function(new_solutions[i]) > fitness_function(sorted_trees[i]) else sorted_trees[i]
```
